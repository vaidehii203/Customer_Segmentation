## **Feature Scaling**

```python
x= df.iloc[:,2:].values
x
```

```python
scale= MinMaxScaler()# creates instance for MinMaxScaler
features=scale.fit_transform(x)# scales the numerical scale in range of 0 to 1
features
```

```python
inertia = []
for i in range(1,16):
    k_means = KMeans(n_clusters=i)
    k_means = k_means.fit(features)
    inertia.append(k_means.inertia_)
```

```python
# elbow graph
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16), inertia, 'bo-')
plt.xlabel('No of clusters') ,plt.ylabel('Inertia')

# elbow visualizer
plt.subplot(1,2,2)
kmeans = KMeans()
visualize = KElbowVisualizer(kmeans,k=(1,16))
visualize.fit(features)
plt.suptitle("Elbow Graph and Elbow Visualizer")
visualize.poof()
plt.show()
```

```python
#silhouette score
silhouette_avg = []
for i in range(2,16):
    # initialise kmeans
    kmeans = KMeans(n_clusters= i)
    cluster_labels = kmeans.fit_predict(features)
    # silhouette score
    silhouette_avg.append(silhouette_score(features,cluster_labels))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhouette_avg,'bX-')
plt.xlabel('Values of K')
plt.ylabel('Silhouette score')
plt.title('Silhouette analysis For Optimal k')
plt.show()
```

## **K-Means Clustering(Take cluster value as 3 as per Elbow method)**

```python
model = KMeans(n_clusters=3)
model = model.fit(features)
```

```python
y_km = model.predict(features)
centers = model.cluster_centers_
```

### Normalized product category interest

```python
product_columns = [col for col in df.columns if col not in ['Cust_ID', 'Gender', 'Orders', 'Cluster']]

# Aggregate total searches for each product category by cluster
cluster_product_interest = df.groupby('Cluster')[product_columns].sum().reset_index()
cluster_product_interest.set_index('Cluster', inplace=True)

#Normalize the data within each cluster
normalized_cluster_product_interest = cluster_product_interest.div(cluster_product_interest.sum(axis=1), axis=0)

#Standardize the data within each cluster
scaler = StandardScaler()
standardized_cluster_product_interest = pd.DataFrame(
    scaler.fit_transform(cluster_product_interest.T).T,
    index=cluster_product_interest.index,
    columns=cluster_product_interest.columns
)



# Normalized Bar Plots
plt.figure(figsize=(20, 15))
for i, cluster in enumerate(normalized_cluster_product_interest.index, 1):
    plt.subplot(len(normalized_cluster_product_interest), 1, i)
    sns.barplot(x=normalized_cluster_product_interest.columns, y=normalized_cluster_product_interest.loc[cluster])
    plt.title(f'Normalized Product Category Interest in Cluster {cluster}')
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
# Standardized Heatmap
plt.figure(figsize=(20, 15))
sns.heatmap(standardized_cluster_product_interest, cmap='viridis', annot=True)
plt.title('Standardized Cluster-Wise Product Category Interest')
plt.show()
```
